# **Exemplo 1**

a) Determine a força de cisalhamento nas interfaces entre as mesas e a alma da seção.
b) Sabendo que o espaçamento entre os pregos é de 5 cm, determine a força de cisalhamento em cada prego.
c) Se o diâmetro do prego for de 3,5 mm, calcule a tensão de cisalhamento em cada prego
