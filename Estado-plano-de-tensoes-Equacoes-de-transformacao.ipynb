# Transformação de tensão
    
### Equações de transformação de tensão
$\sigma_{x'}=\frac{\sigma_{x}+\sigma_{y}}{2} + \frac{\sigma_{x}-\sigma_{y}}{2}\cdot cos2\theta + \tau_{xy}\cdot sen2\theta$

$\sigma_{y'}=\frac{\sigma_{x}+\sigma_{y}}{2} - \frac{\sigma_{x}-\sigma_{y}}{2}\cdot cos2\theta - \tau_{xy}\cdot sen2\theta$

$\tau_{x'y'}=- \frac{\sigma_{x}-\sigma_{y}}{2}\cdot sen2\theta + \tau_{xy}\cdot cos2\theta$
<br>
<br>
   
    
#### Tensão de cisalhamento máxima
$\tau_{máx} = \sqrt{\left(\frac{\sigma_{x}+\sigma_{y}}{2} \right)^2+\tau_{xy}^2}$
    
$\theta_{p}=\frac{1}{2} \cdot arctg \left(-\frac{\sigma_{x} - \sigma_{y}}{2 \theta_{xy}} \right)$
<br>
<br>



#### Tensões principais

$\sigma_{1,2}=\frac{\sigma_{x}+\sigma_{y}}{2}+\sqrt{\left(\frac{\sigma_{x}+\sigma_{y}}{2} \right)^2+\tau_{xy}^2}$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sympy import latex

sigmax = -20 #MPa
sigmay = 9 #MPa
tauxy = 6 #MPa


angulo = np.arange(360); sigmax_1=np.zeros(360)
sigmay_1=np.zeros(360)
tauxy_1=np.zeros(360)
for i in np.arange(360):
    theta=(np.pi/180)*i
    sigmax_1[i]=(sigmax+sigmay)/2 + ((sigmax-sigmay)/2)*np.cos(2*theta) + tauxy*np.sin(2*theta)
    sigmay_1[i]=(sigmax+sigmay)/2 - ((sigmax-sigmay)/2)*np.cos(2*theta) - tauxy*np.sin(2*theta)
    tauxy_1[i]=-((sigmax-sigmay)/2)*np.sin(2*theta) + tauxy*np.cos(2*theta)

sigmamed=(sigmax+sigmay)/2
taumax = (((sigmax-sigmay)/2)**2+tauxy**2)**0.5
sigmamax=sigmamed + taumax
sigmamin=sigmamed - taumax

figure(figsize=(12, 4), dpi=80)    
plt.plot(angulo, sigmax_1, 'blue', label="$\sigma_{x}$")
plt.plot(angulo, sigmay_1, 'red', label="$\sigma_{y}$")
plt.plot(angulo, tauxy_1, 'k', label=r"$\tau_{xy}$")

plt.title("Transformação de tensão no plano")
plt.text(370,taumax, r"$\sigma_{máx} = %s MPa$" %round(sigmamax,2))
plt.text(370,0.8*taumax, r"$\sigma_{mín} = %s MPa$" %round(sigmamin,2))
plt.text(370,0.6*taumax, r"$\sigma_{med} = %s MPa$" %round(sigmamed,2))
plt.text(370,0.4*taumax, r"$\tau_{máx} = %s MPa$" %round(taumax,2))
plt.legend()
plt.xlim(0,360)
plt.xlabel(r"$\theta  (°)$")
plt.ylabel("Tensão (MPa)")
plt.grid()


```

### Conclusões após a análise do gráfico de transformação de tensões no plano
* #### As tensões normais máxima e mínima ocorrem sempre para um mesmo ângulo $\theta$
* #### No ângulo onde ocorrem as tensões normais máxima e mínima, a tensão de cisalhamento é nula
* #### No ângulo onde ocorre as tensões de cisalhamento máxima e mínima, as tensões normais possuem valor igual à sua média: $\sigma_{x}=\sigma_{y}=\sigma_{med} = \frac{\sigma_{x} + \sigma_{y}}{2}$
